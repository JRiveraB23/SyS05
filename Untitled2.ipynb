```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
#periodos y frecuencias
T1 = np.pi/2
T2 = 2*np.pi/5
T3 = np.pi/5
F1 = 1/T1
F2 = 1/T2
F3 = 1/T3
T = 2*np.pi
Fs = 5 * F3
Ts = 1/Fs
t = np.arange(0, 2*T, Ts)
x = -7*np.cos(2*np.pi*F1*t) - 2*np.cos(2*np.pi*F2*t) + 2*np.cos(2*np.pi*F3*t)
```

```python
#cero y pendiente
def ceroypen(x,ymin = -5, ymax = 5):
  xmax = max(x) # x.max()
  xmin = min(x) # x.min()
  m = (ymax - ymin)/(xmax - xmin)
  c = ymin - m*xmin
  y = m*x + c
  return y
y =  ceroypen(x,ymin = -5, ymax = 5)
```

```python
from scipy.spatial.distance import cdist
import numpy as np
#cuantizador
def cuantizador(a, b):
    Ne = b.shape[0]
    dn = cdist(a.reshape(-1, 1), b.reshape(-1, 1))
    ind = np.argmin(dn)
    return b[ind]
```

```python
#n√∫mero de bits
nbits = 3
rmin = -5
rmax = 5
c = np.linspace(rmin,rmax,2**nbits)
#vector cero y pendiente
for i in range(y.shape[0]):
   if i==0:
     b = cuantizador(y[i],c)
   else:
     b = np.append(b,cuantizador(y[i],c))
```
