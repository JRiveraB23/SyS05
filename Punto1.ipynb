## Solucin:

Se tiene que:

$$x(t)=\left|6\sin\left(3t+\frac{\pi}{4}\right)\right|^2 = 6^2\sin^2\left(3t+\frac{\pi}{4}\right) = 36\left(\frac{1}{2}-\frac{\cos\left(6t+\frac{\pi}{2}\right)}{2}\right) = \frac{36}{2}-\frac{36\cos\left(6t+\frac{\pi}{2}\right)}{2} = 18 - 18\cos\left(6t+\frac{\pi}{2}\right)$$

Recordemos que:

$\cos(\theta+\frac{\pi}{2}) = \cos(\theta)\cos(\frac{\pi}{2})-\sin(\theta)\sin(\frac{\pi}{2}) = -\sin(\theta)$

Entonces:

$$\boxed{x(t) = 18+18\sin(6t)}$$

Ahora, por serie trigonomtrica:

$$x(t) = a_0 + \sum_{n=1}^N{a_n\cos(n\omega_0 t) + b_n\sin(n\omega_0 t)},$$

donde:

$$a_0 = c_0 = \frac{1}{t_f-t_i}\int_{t_i}^{t_f}x(t)dt$$

y

$$a_n = \frac{2}{t_f-t_i}\int_{t_i}^{t_f}x(t)\cos(n\omega_0 t)dt$$

$$b_n = \frac{2}{t_f-t_i}\int_{t_i}^{t_f}x(t)\sin(n\omega_0 t)dt$$

Paso 1:

Se comienza con la funcin dada: $x(t) = \left|6\sin\left(3t+\frac{\pi}{4}\right)\right|^2$.

Paso 2:

Se utiliza la identidad trigonomtrica $\sin^2(\theta) = \frac{1-\cos(2\theta)}{2}$ para simplificar la expresin.

Paso 3:

Se sustituye $\theta$ por $3t+\frac{\pi}{4}$ en la identidad trigonomtrica: $\sin^2(3t+\frac{\pi}{4}) = \frac{1-\cos(2(3t+\frac{\pi}{4}))}{2}$.

Paso 4:

Se simplifica la expresin dentro del coseno: $\cos(2(3t+\frac{\pi}{4}))$.

Paso 5:

Se resuelve el coseno utilizando las propiedades trigonomtricas y se simplifica an ms la expresin.

Paso 6:

Se sustituye la expresin simplificada en la identidad trigonomtrica: $\sin^2(3t+\frac{\pi}{4}) = \frac{1-\text{expresin simplificada}}{2}$.

Paso 7:

Se simplifica an ms la expresin si es posible.

Paso 8:

Se obtiene la solucin final de $x(t)$: $x(t) = 18+18\sin(6t)$.

Luego, se procede a encontrar los coeficientes de Fourier utilizando las frmulas correspondientes.

```python
#Importar bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact,IntSlider

#Definir parametros de tiempo y frecuencia
ti = -np.pi #tiempo inicial
tf = np.pi #tiempo final
wo = 2*np.pi/(tf-ti) #frecuencia angular
Fo = wo/2*np.pi #frecuencia fundamental
print('wo=',wo) #Mostrar las frecuencias
print('Fo=',Fo)
Fs = 10*Fo #frecuencia muestreo -> debe respetar Nyquist Fs >= 2 Fo, definir el tiempo de muestreo
tv = np.arange(ti,tf,1/Fs) #vector de tiempo
xo = (np.abs(6*np.sin(3*tv+(np.pi/4))))**2 #seal de entrada original o de entrada
x = abs(6 * np.sin(3 * tv + np.pi/4))**2 #seal de entrada

#Configuracion y visualizacion de la grafica
plt.plot(tv,x,color='green') #grfica de la seal
plt.xlabel('tiempo [s]')
plt.ylabel('$x(t)$')
plt.grid()
plt.show()
```

Dado que $x(t)$ presenta simetra impar, $x(t)=-x(-t)$:


 $$\boxed{a_n = 0 \quad \forall n \in \{1,\dots,N\}}$$

As:

$$x(t)=\left|6\sin\left(3t+\frac{\pi}{4}\right)\right|^2 = 18+18\sin(6t) = a_0 + \sum_{n=1}^N{b_n\sin(n\omega_0 t)}$$

$a_0$ codifica el nivel DC de la seal, entonces:

$$a_0 = \frac{1}{\pi-(-\pi)}\int_{-\pi}^{\pi}\left(18+18\sin(6t)\right)dt = \frac{18}{2\pi}\int_{-\pi}^{\pi}dt+\frac{18}{2\pi}\int_{-\pi}^{\pi}\sin(6t)dt$$

$$a_0 = \frac{18}{2\pi} t \Big|_{-\pi}^{\pi} - \frac{18}{12\pi}\cos(6t)\Big|_{-\pi}^\pi = \frac{18(\pi-(-\pi))}{2\pi}-\frac{18}{12\pi}\left(\cos(6\pi)-\cos(-6\pi)\right) = \frac{18 \cdot 2\pi}{2\pi} - 0 = 18$$

$$\boxed{a_0 = 18}$$

Por otra parte,

$$b_n = \frac{2}{\pi-(-\pi)}\int_{-\pi}^{\pi}x(t)\sin(n\omega_0 t)dt = \frac{2}{2\pi}\int_{-\pi}^{\pi}\left( 18+18\sin(6t)\right)\sin(n\omega_0 t)dt$$

$$b_n = \frac{18}{\pi}\int_{-\pi}^{\pi}{\sin(n\omega_0 t)dt} + \frac{18}{\pi}\int_{-\pi}^{\pi}{\sin(6t)\sin(n\omega_0 t)dt}$$

Teniendo en cuenta que:


$$\sin(\theta)\sin(\alpha)=\frac{\cos(\theta-\alpha)-\cos(\theta+\alpha)}{2}$$

Entonces:

$$b_n = \frac{18}{\pi}\int_{-\pi}^{\pi}{\sin(n\omega_0 t)dt} + \frac{9}{\pi}\int_{-\pi}^{\pi}{\cos([6-n\omega_0] t)dt} - \frac{9}{\pi}\int_{-\pi}^{\pi}{\cos([6+n\omega_0] t)dt}$$

En este caso: $\boxed{\omega_0 = 2\pi/2\pi = 1 [rad/s]}$

$$b_n = -\frac{18}{\pi n }\cos(nt)\Big|_{-\pi}^{\pi} + \frac{9}{\pi(6-n)}\sin([6-n]t)\Big|_{-\pi}^{\pi} - \frac{9}{\pi(6+n)}\sin([6+n]t)\Big|_{-\pi}^{\pi}$$

$$b_n = -\frac{18}{\pi n }\left(\cos(n\pi)-\cos(-n\pi)\right) + 9\frac{\sin([6-n]\pi)-\sin(-[6-n]\pi)}{\pi(6-n)}-9\frac{\sin([6+n]\pi)-\sin(-[6+n]\pi)}{\pi(6+n)}$$

Recordemos que $\cos(\theta)=\cos(-\theta)$ y $\sin(\theta)=-\sin(-\theta)$, entonces:

$$b_n = -\frac{18}{\pi n }\left(\cos(n\pi)-\cos(n\pi)\right) + 9\frac{\sin([6-n]\pi)+\sin([6-n]\pi)}{\pi(6-n)}-9\frac{\sin([6+n]\pi)+\sin([6+n]\pi)}{\pi(6+n)}$$

$$b_n = 18\frac{\sin([6-n]\pi)}{\pi(6-n)}-18\frac{\sin([6+n]\pi)}{\pi(6+n)}$$


Para $n\neq 6,$ $b_n = 0.$ No obstante, para $n = 6$ debemos calcular el lmite y aproximar la indeterminacin $\frac{0}{0}$:


$$b_6 = 18\lim_{n\to 6}\frac{\tfrac{d}{dn}\left[\sin([6-n]\pi)\right]}{\tfrac{d}{dn}\left[\pi(6-n)\right]} = 18\lim_{n\to 6}\frac{\cos([6-n]\pi)(-\pi)}{-\pi}$$

$$b_6 = 18\cos(0)= 18$$

$$\boxed{b_6 = 18}$$


$$b_{-6 }= -18\lim_{n\to {-6}}\frac{\tfrac{d}{dn}\left[\sin([6+n]\pi)\right]}{\tfrac{d}{dn}\left[\pi(6+n)\right]} = -18\lim_{n\to -6}\frac{\cos([6+n]\pi)(\pi)}{\pi}$$

$$b_{-6} = -18\cos(0)= -18$$

$$\boxed{b_{-6} = -18}$$

Por consiguiente:

$$\boxed{b_n = \left\{\begin{align} 0 & \quad \forall n \setminus \{-6,6\}\\
18 & \quad n = 6\\
-18 & \quad n = -6\\
\end{align}\right.}$$

$$\boxed{x(t) = a_0+\sum_{n=1}^{N}{b_n \sin(nt)} = a_0+b_6\sin(6t)=18+18\sin(6t)}$$

Para el caso de la serie exponencial compleja:

$$c_0 = a_0 = 18$$

y

$$c_n = \frac{a_n - j b_n}{2}$$

Entonces:

$$\boxed{c_n = \left\{\begin{align}
9j & \quad n = -6\\
18 & \quad n = 0\\
-9j & \quad n = 6\\
0 & \quad \forall n \setminus \{-6,0,6\}
\end{align}\right.}$$

y

$$\boxed{x(t) = \sum_{n=-N}^{N}{c_n e^{j n t}}}$$


$$x(t) = c_{-6}e^{-j6t} + c_{0}e^{0} + c_{6}e^{j6t}$$

$$x(t)= 9j(\cos(6t)-j\sin(6t))+18-9j(\cos(6t)+j\sin(6t)) $$

$$\boxed{x(t) = 9j\cos(6t)+9\sin(6t))+18-9j\cos(6t)+9\sin(6t)=18+18\sin(6t)}$$


*Paso 1:* Dada la simetra impar de \(x(t)\), se concluye que \(x(t) = -x(-t)\), lo que implica que todos los coeficientes \(a_n\) son iguales a cero.

*Paso 2:* Se presenta la expresin general de la seal \(x(t)\) en trminos de la serie de Fourier \(a_0 + \sum_{n=1}^N{b_n\sin(n\omega_0 t)}\).

*Paso 3:* Se calcula \(a_0\), el cual representa el nivel DC de la seal. La integral se evala utilizando propiedades trigonomtricas y se llega a la conclusin de que \(a_0 = 18\).

*Paso 4:* Se calculan los coeficientes \(b_n\) utilizando la frmula \(b_n = \frac{2}{\pi}\int_{-\pi}^{\pi}x(t)\sin(n\omega_0 t)dt\). Se aplican identidades trigonomtricas para simplificar la integral y se obtiene \(b_n = 18\frac{\sin([6-n]\pi)}{\pi(6-n)}-18\frac{\sin([6+n]\pi)}{\pi(6+n)}\).

*Paso 5:* Se determina que \(b_n = 0\) para \(n\) diferente de \(-6\) y \(6\). Para \(n = 6\), se calcula \(b_6 = 18\) y para \(n = -6\), se calcula \(b_{-6} = -18\).

*Paso 6:* Se presenta la expresin final para \(b_n\) en trminos de \(n\), indicando que \(b_n = 18\) para \(n = 6\), \(b_{-6} = -18\), y \(b_n = 0\) para \(n\) diferente de \(-6\) y \(6\).

*Paso 7:* Se reconstruye la seal utilizando la serie de Fourier y se llega a la conclusin de que \(x(t) = 18+18\sin(6t)\).

*Paso 8:* Se presenta la serie exponencial compleja y se calculan los coeficientes \(c_n\) utilizando la relacin \(c_n = \frac{a_n - j b_n}{2}\). Se concluye que \(c_n = 9j\) para \(n = -6\), \(c_n = 18\) para \(n = 0\), y \(c_n = -9j\) para \(n = 6\).

*Paso 9:* Se presenta la expresin final para \(c_n\) en trminos de \(n\).

*Paso 10:* Se reconstruye la seal utilizando la serie exponencial compleja y se llega a la conclusin de que \(x(t) = 18+18\sin(6t)\).

```python
#Se crean las bases de la Serie Exponencial Compleja de Fourier.
#Definicion de los parametros iniciales
T=tf-ti #Intervalo de inters
N=20 #Nmero armonicos a simular
wo=2*np.pi/T #Frecuencia fundamental
#Definimos las bases:
phin=np.zeros((len(tv), 2*N+1), dtype=np.complex_) #Crear matriz para guardar bases. Bases de la serie exponencial compleja de fourier
for n in range(-N, N+1, 1):
    phin[:, n+N]=np.exp(1j*n*wo*tv) #Base Exponencial Compleja de Fourier en el intrvalo de inters.
#Para graficar las bases:
def pltbase(n=1):
    plt.plot(tv, np.real(phin[:, int(n+N)]), label="Re{$\phi_n$(t)}")
    plt.xlabel("t[s]", fontsize = 14)
    plt.ylabel("$\phi_n$(t)", fontsize= 14)
    plt.plot(tv, np.imag(phin[:, int(n+N)]), label="Im{$\phi_n$(t)}")
    plt.grid()
    plt.legend()
    plt.show()
    return
interact(pltbase, n=(-N,N,1))
plt.show()
```

```python
#Calcular el espectro y cambiar segn la seal estudiada:
cn=np.zeros(2*N+1, dtype=np.complex_)
nv=np.linspace(-N, N, 2*N+1) #Vector de armnicos
cn[N]=18 #Nivel DC de C0
cn[N-6]=9j #Armnico -6
cn[N+6]=-9j #Armnico 6
#Graficar espectro:
fig=plt.figure(figsize=(8, 8))
plt.subplot(2, 2, 1)
plt.stem(nv, np.real(cn), 'r') #Grfica de la parte real del espectro
plt.xlabel(r'$nw_o$[rad/s]', fontsize=14)
plt.ylabel(r'$Re\{C_n\}$', fontsize=14)
plt.axis('tight')
plt.grid()
plt.subplot(2, 2, 2)
plt.stem(nv, np.imag(cn), 'r') #Grafica de la parte imaginaria del espectro
plt.xlabel(r'$nw_o$[rad/s]', fontsize=14)
plt.ylabel(r'$Im\{C_n\}$', fontsize=14)
plt.axis('tight')
plt.grid()
plt.subplot(2, 2, 3)
plt.stem(nv, abs(cn), 'r') #Magnitud del espectro
plt.xlabel(r'$nw_o$[rad/s]', fontsize=14)
plt.ylabel(r'$|C_n|$', fontsize=14)
plt.axis('tight')
plt.grid()
plt.subplot(2, 2, 4)
plt.stem(nv, np.angle(cn), 'r') #Fase del espectro
plt.xlabel(r'$nw_o$[rad/s]', fontsize=14)
plt.ylabel(r'$\langle C_n$', fontsize=14)
plt.axis('tight')
plt.grid()
fig.tight_layout()
plt.show()
```

```python
Px=486 #Potencia de la seal x(t)
def pltest(Na=1):
    ind=range(N-Na, N+Na+1)
    er=1-np.sum(abs(cn[ind])**2)/Px
#Seal reconstruida:
    xe=phin[:, ind].dot(cn[ind]) #dot calcula el producto punto entre vectores
    plt.plot(tv, xe, color='b', label='$x_e(t)$') #Seal estimada o filtrada
    plt.plot(tv, x, color='r', label='$x(t)$') #Seal original
    plt.title('ER=%.2f[%%]'%(100*er), fontsize=16)
    plt.xlabel('t[s]')
    plt.ylabel('x(t)')
    plt.grid()
    plt.legend()
    plt.show()
    return
interact(pltest, Na=(1, N,1))
plt.show()
```
